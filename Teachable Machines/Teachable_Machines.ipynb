```python
# Import required libraries
import tensorflow as tf
import tensorflowjs as tfjs
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python

# Load the model (model.json and weights.bin must be in same directory)
model = tfjs.converters.load_keras_model('/content/model.json')

# Show model structure
model.summary()
```

```python
# Define labels in the same order used during training
labels = ['book', 'mug']  # Ensure this matches your Teachable Machine training order

# Image preprocessing function
def preprocess_image(img_path):
    img = Image.open(img_path).resize((224, 224)).convert('RGB')
    img_array = np.array(img) / 255.0
    return np.expand_dims(img_array, axis=0)
```

```python
# Prediction function
def predict_and_display(image_path):
    image_tensor = preprocess_image(image_path)
    predictions = model.predict(image_tensor)
    predicted_label = labels[np.argmax(predictions)]
    confidence = np.max(predictions)

    plt.imshow(Image.open(image_path))
    plt.title(f"Predicted: {predicted_label} (Confidence: {confidence:.2f})")
    plt.axis('off')
    plt.show()

    print(f"Predicted Label: {predicted_label}, Confidence: {confidence:.2f}")
```

```python
# Example: Predict on a book or mug image
predict_and_display('/content/test1.jpg')
```

```python
# Step 2: Import required libraries
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import cv2
import os
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
# Step 3: Load MoveNet Lightning from TensorFlow Hub
movenet = hub.load("https://tfhub.dev/google/movenet/singlepose/lightning/4")
```

```python
# Step 4: Function to extract keypoints (flattened 51 values)
def detect_pose(img_path):
    img = cv2.imread(img_path)
    if img is None:
        raise FileNotFoundError(f"Image not found: {img_path}")
    img = cv2.resize(img, (192, 192))
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    input_tensor = tf.expand_dims(tf.cast(rgb, tf.int32), axis=0)
    keypoints = movenet.signatures['serving_default'](input=input_tensor)['output_0'].numpy()
    return keypoints.flatten().astype(np.float32)
```

```python
pose_images = {
    'One Hand Raised': [
        '/content/h1.JPG', '/content/h2.JPG', '/content/h3.JPG',
        '/content/h4.JPG', '/content/h5.JPG', '/content/h6.JPG',
        '/content/h7.JPG', '/content/h8.JPG', '/content/h9.JPG',
        '/content/h10.JPG'
    ],
    'Both Hands Raised': [
        '/content/h21.JPG', '/content/h22.JPG', '/content/h23.JPG',
        '/content/h24.JPG', '/content/h25.JPG', '/content/h26.JPG'
    ]
}
```

```python
# Label mapping
label_map = {'One Hand Raised': 0, 'Both Hands Raised': 1}

# Extract keypoints and create dataset
X, y = [], []

for label, filepaths in pose_images.items():
    for filepath in filepaths:
        print(f"Processing: {filepath}")
        keypoints = detect_pose(filepath)
        X.append(keypoints)
        y.append(label_map[label])

X = np.array(X)
y = np.array(y)

print("Dataset created successfully.")
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
import matplotlib.pyplot as plt
import cv2

# Display all images from the dataset
for label, filepaths in pose_images.items():
    print(f"\nShowing images for class: {label}")
    for filepath in filepaths:
        img = cv2.imread(filepath)
        if img is not None:
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.imshow(img_rgb)
            plt.title(f"{label} - {os.path.basename(filepath)}")
            plt.axis('off')
            plt.show()
        else:
            print(f"Image not found or failed to load: {filepath}")
```

```python
# Step 6: Train logistic regression classifier
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X, y)
print("Classifier trained successfully.")

# Evaluate training accuracy
accuracy = classifier.score(X, y)
print(f"Training Accuracy: {accuracy:.2f}")
```

```python
# Step 7: Predict on a new test image
def predict_pose(image_path):
    keypoints = detect_pose(image_path).reshape(1, -1)
    prediction = classifier.predict(keypoints)[0]
    label_inv = {v: k for k, v in label_map.items()}
    predicted_label = label_inv[prediction]

    # Show result
    img = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(f"Predicted: {predicted_label}")
    plt.axis('off')
    plt.show()

    print(f"Predicted Pose: {predicted_label}")
```

```python
# Run prediction on any image
predict_pose('/content/test6.jpg')
```

```python
# Run prediction on any image
predict_pose('/content/test7.jpg')
```

```python
# Step 2: Import packages
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Step 3: Load pre-trained Universal Sentence Encoder from TF Hub
use_model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
print("Universal Sentence Encoder is ready.")
```

```python
# Step 4: Sample text data (Neutral vs Opinionated sentiment)
texts = [
    "The event was held on Sunday evening.",        # Neutral
    "This service is absolutely fantastic!",         # Opinionated
    "We met at the cafe and discussed logistics.",   # Neutral
    "Totally love the new features, well done!",     # Opinionated
]

categories = ["Neutral", "Opinionated", "Neutral", "Opinionated"]

# Convert categories to numeric labels
label_encoder = {"Neutral": 0, "Opinionated": 1}
y = np.array([label_encoder[label] for label in categories])

print("Text samples and labels encoded.")
```

```python
# Step 5: Convert text to embeddings using USE
X = use_model(texts).numpy()
print("USE Embeddings shape:", X.shape)
```

```python
# Step 6: Train a logistic regression classifier
model = LogisticRegression()
model.fit(X, y)

# Predict on training data
predicted = model.predict(X)

# Show classification report
inverse_labels = ["Neutral", "Opinionated"]
report = classification_report(y, predicted, target_names=inverse_labels)
print("Model performance on training data:\n")
print(report)
```

```python
# test samples
test_sentences = [
    "He arrived at 5 PM and left by 7.",              # Neutral
    "I absolutely hate how slow this app is.",        # Opinionated
    "Lunch was served in the conference room.",       # Neutral
    "This product exceeded all my expectations!",     # Opinionated
    "We discussed the meeting schedule and agenda.",  # Neutral
    "Highly recommend this course to everyone!",      # Opinionated
]

# Expected labels (for manual verification if needed)
expected_labels = ["Neutral", "Opinionated", "Neutral", "Opinionated", "Neutral", "Opinionated"]

# Convert to embeddings
X_test = use_model(test_sentences).numpy()

# Predict
test_predictions = model.predict(X_test)

# Map back to label names
inv_map = {v: k for k, v in label_encoder.items()}
predicted_labels = [inv_map[p] for p in test_predictions]

# Display results
for i, sentence in enumerate(test_sentences):
    print(f"\nSentence: {sentence}")
    print(f"Predicted: {predicted_labels[i]} | Expected: {expected_labels[i]}")
```
